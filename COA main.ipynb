```python
print(" 1. Converts Binary to Gray code ")
print(" 2. Converts Gray code to Binary ")
print(" 3. Converts Binary to Decimal ")
print(" 4. Converts Decimal to Binary ")
print(" 5. Converts Binary to BCD ")
print(" 6. Converts Binary to Excess-3 ")
print(" 7. Converts Binary to Hexadecimal ")
print(" 8. Converts Binary to Octal ")
user_input = int(input("Enter a number: "))
if user_input == 1:
# Converts Binary to Gray code
    def binary_to_gray(n) :
        n=int(n,2)
        n^=(n>>1)
        return bin(n)[2:]
    g= input("Enter Binary Number: ")
    b= binary_to_gray(g)
    print("Gray Codeword: ",b)
    
elif user_input == 2:
# Converts Gray code to Binary
    def gray_to_binary(n) :
        n=int(n,2)
        mask=n
        while mask!=0:
            mask>>=1
            n^=mask
        return bin(n) [2:]
    g= input("Enter Gray Codeword: " )
    b= gray_to_binary(g)
    print ("In Binary: ",b)
     
elif user_input == 3:
# Converts Binary to Decimal
     b= input("Enter Binary Number: ")
     d=int(b,2)
     print("In Decimal Number: ",d)
    
elif user_input == 4:
# Converts Decimal to Binary
    def decimal_to_binary(n):
        binary = ""
        while n > 0:
            bit = str(n % 2)
            binary = bit + binary
            n = n // 2
        return binary
    decimal_num = int(input("Enter a decimal number: "))
    binary_num = decimal_to_binary(decimal_num)
    print("In Binary: ",binary_num)
    
elif user_input == 5:
# Converts Binary to BCD
    def binary_to_bcd(binary):
        decimal = 0
        for i, bit in enumerate(reversed(binary)):
            decimal += int(bit) * (2 ** i)
        bcd = ''
        while decimal > 0:
            digit = decimal % 10
            bcd_digit = ''
            for i in range(4):
                bcd_digit = str(digit % 2) + bcd_digit
                digit //= 2
            bcd = bcd_digit + bcd
            decimal //= 10
        return bcd
    binary_num = input("Enter a binary number: ")
    bcd_num = binary_to_bcd(binary_num)
    print("In BCD : ",bcd_num)

elif user_input == 6:
# Converts Binary to Excess-3 
    def binary_to_excess3(binary):
        decimal = int(binary, 2)
        excess3 = decimal + 3
        excess3_binary = bin(excess3)[2:]  
        return excess3_binary.zfill(4) 
    binary_num = input("Enter a binary number: ")
    excess3_code = binary_to_excess3(binary_num)
    print("In Excess-3: ",excess3_code)

elif user_input == 7:
# Converts Binary to Hexadecimal
    def binary_to_hex(binary):
        decimal = int(binary, 2)
        hex_value = hex(decimal)[2:] 
        return hex_value.upper() 
    binary_num = input("Enter a binary number: ")
    hexadecimal = binary_to_hex(binary_num)
    print("In Hexadecimal: ",hexadecimal)

elif user_input == 8:
# Converts Binary to Octal
    def binary_to_octal(binary):
        decimal = int(binary, 2)
        octal_value = oct(decimal)[2:] 
        return octal_value
    binary_num = input("Enter a binary number: ")  
    octal = binary_to_octal(binary_num)
    print("In Octal: ",octal)
    
else:
# Code to run if the user enters anything else
     print("Invalid input")
```
